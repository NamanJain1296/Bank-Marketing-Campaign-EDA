# Bar Graphs 

```python
job_counts = b['job'].value_counts()
ax = job_counts.plot(kind = 'barh', color = 'skyblue', edgecolor='black')
for index, value in enumerate(job_counts):
    ax.text(value, index, str(value))
    
plt.xlabel('Count')
plt.ylabel('Job')
plt.title('Distribution fo Jobs')

plt.show()
```

- The bar graph shows the count of people various jobs approached by bank. The job of 'admin' recieved maximum calls of all.

```python
marry = b['marital'].value_counts()
ax = marry.plot(kind = 'barh', color = 'yellow', edgecolor='black')
for index, value in enumerate(marry):
    ax.text(value, index, str(value))
    
plt.xlabel('Count')
plt.ylabel('Marital')
plt.title('Distribution for Marital Status')

plt.show()
```

- The bar graph show the count for people depicting their marital status. From the data, 19953 people were married, 9257 were single, 3675 were divorced and record for 65 was unknown.

```python
edu = b['education'].value_counts()
ax = edu.plot(kind = 'barh', color = 'red', edgecolor='black')
for index, value in enumerate():
    ax.text(value, index, str(value))
    
plt.xlabel('Count')
plt.ylabel('Education Status')
plt.title('Distribution for Education Status')

plt.show()
```

```python
mon = b['month'].value_counts()
ax = mon.plot(kind = 'barh', color = 'pink', edgecolor='black')
for index, value in enumerate(mon):
    ax.text(value, index, str(value))
    
plt.xlabel('Count')
plt.ylabel('Month')
plt.title('Distribution for Months')

plt.show()
```

# Pie Chart

```python
house = b['housing'].value_counts()
house.plot.pie(autopct = '%1.1f%%', startangle = 90, figsize = (8,8), title = 'Each Response')
```

- The data has 52.4% yes for housing, 45.2% no for housing and the remaining for unkown.


```python
loans = b['loan'].value_counts()
loans.plot.pie(autopct = '%1.1f%%', startangle = 90, figsize = (8,8), title = 'Each Response')
```

- The data shows that 82.3% did not take the loan, 15.2% took the loan and data for about 2.4% people are unknown.

```python
contacts = b['contact'].value_counts()
contacts.plot.pie(autopct = '%1.1f%%', startangle = 90, figsize = (8,8), title = 'Each Response')
```

- The data shows that 63.5% people gave their cellular number and 36.5% people gave their telephone number. 

```python
week = b['day_of_week'].value_counts()
week.plot.pie(autopct = '%1.1f%%', startangle = 90, figsize = (8,8), title = 'Days of Week')
```

- The data shows that the maximum calls were made on Thursday i.e 20.8% followed by Monday(20.7%) and so on.

```python
prev = b['previous'].value_counts()
prev.plot.pie(autopct = '%1.1f%%', startangle = 90, figsize = (8,8), title = 'Previous Values')
```

```python
p = b['poutcome'].value_counts()
p.plot.pie(autopct = '%1.1f%%', startangle = 90, figsize = (8,8), title = 'P-Outcomes')
```

- The data shows the outcomes for various calls made by the bank. 86.2% are nonexistent, 10.4% are failures and only 3.4% are success.
